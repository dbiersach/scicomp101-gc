**Import needed packages**

```python
# Cell 1
import matplotlib.pyplot as plt
import numpy as np
```

**Create 10,000 random radii between [0, 1) - <mark>using the sqrt() operator!</mark>**

```python
# Cell 2
n = 10_000
r = np.sqrt(np.random.rand(n))
```

**Create 10,000 angles equally spaced between $0\le\theta\le 2\pi$**

```python
# Cell 3
theta = np.linspace(0, 2 * np.pi, n)
```

**Create Cartesian points $(x,y)$ based upon polar coordinate conversion from $(r,\theta)$**

```python
# Cell 4
x = r * np.cos(theta)
y = r * np.sin(theta)
```

**Plot all points $(x,y)$**

```python
# Cell 5
plt.scatter(x, y, s=0.5)
plt.title("Uniform Circle Sampling")
plt.gca().set_aspect("equal")
plt.show()
```
