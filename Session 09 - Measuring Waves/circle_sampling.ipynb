**Import needed packages**

```python
# Cell 1
import matplotlib.pyplot as plt
import numpy as np
```

**Create 10,000 random radii between [0, 1)**

```python
# Cell 2
n = 10_000
r = np.random.rand(n)
```

**Create 10,000 angles equally spaced between $0\le\theta\le 2\pi$**

```python
# Cell 3
theta = np.linspace(0, 2 * np.pi, n)
```

**Create Cartesian points $(x,y)$ based upon polar coordinate conversion from $(r,\theta)$**

```python
# Cell 4
x = r * np.cos(theta)
y = r * np.sin(theta)
```

**Plot all points $(x,y)$**

```python
# Cell 5
plt.scatter(x, y, s=0.5)
plt.title("Uniform Circle Sampling")
plt.gca().set_aspect("equal")
plt.show()
```

**The above graph is *NOT* good - it is not <u>uniform</u> `(fair)` sampling - how can we fix it?** \
Hint: consider the graph below - what happens we we take the square root of the $y$ values?

```python
# Cell 6
x = np.random.rand(n)
y1 = np.random.rand(n)
y2 = np.sqrt(y1)

_, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].scatter(x, y1, s=0.5)
ax[0].set_title("$y=rand$")

ax[1].scatter(x, y2, s=0.5)
ax[1].set_title(r"$y=\sqrt{\left( rand\right)}$")

plt.show()
```
