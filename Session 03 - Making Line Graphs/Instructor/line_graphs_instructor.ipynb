**Import matplotlib and numpy**

```python
# Cell 1
import matplotlib.pyplot as plt
import numpy as np
```

**Create an array x spanning $-10\leq x\leq 10$**

```python
# Cell 2
x = np.linspace(-10, 10)
print(x)
```

**Set $y_1=2x-5$**

```python
# Cell 3
y1 = 2 * x - 5
print(y1)
```

**Set $y_2=-0.3x^2+15$**

```python
# Cell 4
y2 = -0.3 * x**2 + 15
print(y2)
```

**Create one graph that plots both $y1(x)$ and $y2(x)$**

```python
# Cell 5
plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
```
