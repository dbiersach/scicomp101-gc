**Decipher the "bigram_ciphertext.txt" file**
1. Open the file for binary reading and read the contents into a `bytearray`
2. Apply the XOR-128 operation to each byte in the `bytearray`
3. Convert the `bytearray` back to a Python UNICODE string
4. Print the plain text string in rows of 120 characters per row

```python
# Cell 3

file_name = "bigram_ciphertext.txt"
file_path = notebook_path / file_name
with Path.open(file_path, "rb") as f_in:
    f_bytes = bytearray(f_in.read())

for i in range(0, len(f_bytes)):
    f_bytes[i] = f_bytes[i] ^ 128

plain_text = f_bytes.decode()
for i in range(0, len(plain_text), 120):
        print(plain_text[i:i + 120])
```
