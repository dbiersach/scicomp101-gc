**Define a function that Caesar decrypts a file using a shift key value**

```python
# Cell 3
def caesar_decrypt(file_name="", key_shift=0):
    file_path = notebook_path / file_name
    with Path.open(file_path, "rb") as f_in:
        f_bytes = bytearray(f_in.read())

    for i in np.arange(len(f_bytes)):
        f_bytes[i] = (f_bytes[i] + key_shift) % 256

    print(f_bytes.decode("utf-8", "ignore"))
```

**Decrypt the file "ciphertext1.txt"**

```python
# Cell 4
caesar_decrypt(file_name="ciphertext1.txt", key_shift=-45)
```
