**Set string $s_1$ to "Forever Young"**

```python
# Cell 1
s1 = "Forever Young"
s1
```

**Create string $s_2$ by walking backwards through $s_1$ and <u>appending</u> each character in $s_1$ to the *end* of $s_2$**

```python
# Cell 2
s2 = ""
for i in range(len(s1) - 1, -1, -1):
    s2 = s2 + s1[i]
s2
```

**Create string $s_3$ by walking forwards through $s_1$ and <u>prepending</U> each character in $s_1$ to the *front* of $s_3$**

```python
# Cell 3
s3 = ""
for c in s1:
    s3 = c + s3
s3
```

**Use Python `array slicing` syntax to print $s_1$ in reverse order**

```python
# Cell 4
s1[::-1]
```
