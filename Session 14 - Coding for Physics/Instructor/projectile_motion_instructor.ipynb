**Import needed packages/modules**

```python
# Cell 1
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import Rectangle
```

**Specify the simulation constants**
1. The trampoline is 30m from the cannon
2. The cannon is at a fixed launch angle of $45Â°$
3. The acceleration due to gravity is $9.81\ \frac{m}{s^2}$

```python
# Cell 2
rng = 30  # m
t = np.radians(45)  # 45 degree launch angle
g = 9.81  # m/s^2
```

**Set the initial launch velocity $v_0$ in $(m/s)$**\
$v_0=\large\sqrt{\frac{Range\ \times g}{\sin{2\theta}}}$

```python
# Cell 3
v0 = np.sqrt(rng * g / np.sin(2 * t))
```

**Create a linear space for the independent variable that spans $0\le x\le 40$**

```python
# Cell 4
x = np.linspace(0, 40)
display(x)
```

**Create an array for the dependent variable $y(x)$ from the equation of motion:**\
$y=\tan{\left(\theta\right)}\times x-{\large{\frac{g}{2\times v_0^2\times\cos^2{\left(\theta\right)}}}}\times x^2$

```python
# Cell 5
y = np.tan(t) * x - (g / (2 * v0**2 * np.cos(t) ** 2)) * x**2
display(y)
```

**Plot the trajectory of the circus performer**

```python
# Cell 6
plt.plot(x, y)
plt.title("Circus Performer Projectile Motion")
plt.xlabel("Distance (m)")
plt.ylabel("Height (m)")
plt.xlim(left=0)
plt.ylim(bottom=0)
ax = plt.gca()
ax.add_patch(Rectangle((27.5, 0), 5, 1, color="red"))
ax.set_aspect("equal")
plt.show()
```
