**Create a linear space spanning $1\le x \le 10$ having 10 equally spaced elements**

```python
# Cell 1
import numpy as np
x = np.linspace(1, 10, 10)
print(f"{x = }")
```

**Set $y_1$ to the *cummulative sum* of every element in x**\
$y_1 = \sum\limits_{k=1}^{n} x_k$

```python
# Cell 2
y1 = np.cumsum(x)
print(f"{y1 = }")
```

**Set $y_2$ to the result of applying the *Gaussian Summation* Operator to each value in x**\
$y_2=\frac{x_k(x_k+1)}{2}$

```python
# Cell 3
y2 = x * (x + 1) / 2
print(f"{y2 = }")
```

**Does every element in $y_1=y_2$ ?**

```python
# Cell 4
print(np.array_equal(y1, y2))
```
