**Import needed packages/modules**

```python
# Cell 1
from math import floor, sqrt

from numba import njit
```

**Define a `numba` accelerated function to solve Pell's equation:&nbsp;$x^2-n\times y^2=1$**
1. For this example, only search for $2\le n\le 70$ and $1\le x\le 70,000$
2. We only need to check for $y\le\left\lfloor{\sqrt\frac{x^2-1}{n}}\right\rfloor$, <mark>why?</mark>
3. Stop searching upon finding the first solution where $n\in\mathbb{Z}^{+}$
4. If no solution can be found return $(0,0)$

```python
# Cell 2
@njit
def pell_solution(n):
    x = 1
    while x < 70_000:
        x2 = x * x
        y = 1
        y_max = floor(sqrt((x2 - 1) / n))
        while y <= y_max:
            y2 = y * y
            if x2 - n * y2 == 1:
                return True, x, y
            y += 1
        x += 1
    return False, 0, 0
```

**Display the solutions to Pell's equation for $2\le n\le 70$**

```python
# Cell 3
print(f"{'n':>4}{'x':>8}{'y':>8}")
print(f"{'='*3:>4}{'='*6:>8}{'='*6:>8}")

for i in range(2, 71):
    solution_exists, xs, ys = pell_solution(i)
    if solution_exists:
        print(f"{i:>4}{xs:>8}{ys:>8}")
    else:
        print(f"{i:>4}{'-':>8}{'-':>8}")
```
