**Define a function `sqrt(x)` that uses Newton's Method**

```python
# Cell 2
def sqrt(x):
    low, high = 0, x
    est = (low + high) / 2
    while np.abs(est**2 - x) > 1e-10:
        if est**2 < x:
            low = est
        else:
            high = est
        est = (low + high) / 2
    return est
```

**Use `sqrt(x)` to calculate $\sqrt{168923.74}$**

```python
# Cell 3
sqrt(168923.74)
```
