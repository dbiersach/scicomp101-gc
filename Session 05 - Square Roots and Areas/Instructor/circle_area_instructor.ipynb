**Import common packages**

```python
# Cell 1
import numpy as np
```

**Set $n$ to one million (intervals)**

```python
# Cell 2
n = 1_000_000
```

**Create a linear space $0\le x\le 1$ with $n$ intervals**

```python
# Cell 3
x = np.linspace(0, 1, n)
x
```

**Create array $y=\sqrt{1-x^2}$**

```python
# Cell 4
y = np.sqrt(1 - x**2)
y
```

**Sum the area of all the rectangles under the curve**

```python
# Cell 5
a = np.sum(y) / n
a
```

**Multiply area by 4 to estimate the area of entire unit circle**\
$F(x)=4\int\limits_{0}^{1} \sqrt{1-x^2}\,dx$

```python
# Cell 6
a * 4
```
