**Import the scipy.special module to gain access to the `factorial` function**

```python
# Cell 1
import numpy as np
import scipy.special
```

**Create an integer array where $0\le x\le 20$**

```python
# Cell 2
x = np.arange(20)
x
```

**Set $z=0+\pi i$**

```python
# Cell 3
z = complex(0, np.pi)
z
```

**Create array $n=z^x$**

```python
# Cell 4
n = np.power(z, x)
n
```

**Create array $d=x!$**

```python
# Cell 5
d = scipy.special.factorial(x)
d
```

**Divide every term in array $n$ by the corresponding term in array $d$**

```python
# Cell 6
n / d
```

**Calculate $e^z=\sum\limits_{x=0}^{\infty}\frac{z^x}{x!}$ where $z=\pi i$**

```python
# Cell 7
ez = np.sum(n / d)
np.round(ez, 8)
```
