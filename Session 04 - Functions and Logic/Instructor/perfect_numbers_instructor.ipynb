**Import common packages**

```python
# Cell 1

import numpy as np
```

**Define a function to test if an integer $n$ is perfect**

```python
# Cell 2

def is_perfect(n):
    x = np.arange(1, n)
    factors = x[np.where(n % x == 0)]
    return np.sum(factors) == n

print(is_perfect(6))
print(is_perfect(12))
```

**Define a `main()` function to test every integer $n$ where $2\leq n< 10,0000$**

```python
# Cell 3

def main():
    for n in range(2, 10000):
        if is_perfect(n):
            print(n)

main()
```
