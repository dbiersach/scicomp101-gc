**Import common packages**

```python
# Cell 1
import matplotlib.pyplot as plt
import numpy as np
```

**Create an array $t$ that is a linear space spanning $0\le t\le 4\pi$ having 1000 intervals**

```python
# Cell 2
t = np.linspace(0, 4 * np.pi, 1000)
```

**Set $r1=4+4\cos(4t)$**\
**Set $r2=3+3\cos(4t+\pi)$**\
**Set $r3=5+5\cos(\frac{3}{2}t)$**\
**Set $r4=7+7\sin(11t)\cos(5t)$**

```python
# Cell 3
r1 = 4 + 4 * np.cos(4 * t)
r2 = 3 + 3 * np.cos(4 * t + np.pi)
r3 = 5 + 5 * np.cos(3 / 2 * t)
r4 = 7 + 7 * np.sin(11 * t) * np.cos(5 * t)
```

**Plot the $r_1, r_2, r_3$ graphs using a `Polar Projection`**

```python
# Cell 4
plt.subplot(projection="polar")
plt.plot(t, r1)
plt.plot(t, r2)
plt.plot(t, r3)
plt.show()
```

**Plot the $r_4$ superposition using a `Polar Projection`**

```python
# Cell 5
plt.subplot(projection="polar")
plt.plot(t, r4, color="black")
plt.show()
```
